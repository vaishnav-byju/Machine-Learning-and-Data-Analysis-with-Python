## Step 1: Create Dummy Dataset

```python
import pandas as pd

# Create a dummy dataset
data = {
    "Patient_ID": ["P1", "P2", "P3", "P4", "P5"],
    "Gender": ["M", "F", "F", "M", "F"],
    "AppointmentType": ["Routine", "Emergency", "Routine", "Follow-up", "Emergency"],
    "City": ["New York", "Los Angeles", "Chicago", "New York", "Chicago"]
}

df = pd.DataFrame(data)
print("Original Dataset:\n", df)
```

## Step 2: Inspect Dataset

```python
df.info()
df.head()
```

## Step 3: Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Copy dataset
df_label = df.copy()

# Apply Label Encoding to categorical columns
for col in ["Gender", "AppointmentType", "City"]:
    le = LabelEncoder()
    df_label[col] = le.fit_transform(df_label[col])

print("After Label Encoding:\n", df_label)
```

## Step 4: One‑Hot Encoding

```python
# One-Hot Encoding
df_onehot = pd.get_dummies(df, columns=["Gender", "AppointmentType", "City"])

print("After One-Hot Encoding:\n", df_onehot)
```
