### Step 4: Encode Categorical Variables
We convert categorical columns into numerical format using Pandas encoding techniques.


```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Reload dataset (if needed)
df = pd.read_csv("KaggleV2-May-2016.csv")

# --- Fix invalid values ---
# Replace negative ages with median
median_age = df['Age'].median()
df.loc[df['Age'] < 0, 'Age'] = median_age

# Simplify Handcap column (convert >0 to 1)
df['Handcap'] = df['Handcap'].apply(lambda x: 1 if x > 0 else 0)

# --- Encode categorical variables ---
# Encode Gender (F/M → 0/1)
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

# Encode No-show (No/Yes → 0/1)
df['No-show'] = df['No-show'].map({'No':0, 'Yes':1})

# Encode Neighbourhood with LabelEncoder
df['Neighbourhood'] = LabelEncoder().fit_transform(df['Neighbourhood'])

# Convert ScheduledDay and AppointmentDay to datetime
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

print("Dataset Info after Encoding:")
print(df.info())
print("\nFirst 5 rows after encoding:\n", df.head())
```
