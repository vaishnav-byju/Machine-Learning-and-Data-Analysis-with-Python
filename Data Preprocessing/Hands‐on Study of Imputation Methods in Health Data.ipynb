## Step 1: Create Dummy Dataset

```python
import pandas as pd
import numpy as np

# Create a dummy dataset
data = {
    "Patient_ID": ["P1","P2","P3","P4","P5","P6","P7","P8","P9","P10"],
    "Age": [25, 30, np.nan, 22, 40, np.nan, 29, 31, 28, np.nan],
    "Blood_Pressure": [120, 130, 125, np.nan, 140, 135, np.nan, 128, 132, 130],
    "Gender": ["M", "F", "F", "M", np.nan, "F", "M", "M", np.nan, "F"],
    "Cholesterol": [200, 220, 210, 205, np.nan, 215, 225, np.nan, 230, 210]
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)
```

## Step 2: Inspect Dataset

```python
# Check data types and missing values
df.info()

# Get statistical summary of numerical columns
df.describe()
```

## Step 3: Mean Imputation

```python
# Make a copy of the dataset
df_mean = df.copy()

# Apply mean imputation for numerical columns
df_mean.fillna({
    "Age": df["Age"].mean(),
    "Blood_Pressure": df["Blood_Pressure"].mean(),
    "Cholesterol": df["Cholesterol"].mean()
}, inplace=True)

print("After Mean Imputation:\n", df_mean)

# Check summary statistics again
df_mean.describe()
```

## Median Imputation

```python
# Median Imputation
df_median = df.copy()

df_median.fillna({
    "Age": df["Age"].median(),
    "Blood_Pressure": df["Blood_Pressure"].median(),
    "Cholesterol": df["Cholesterol"].median()
}, inplace=True)

print("After Median Imputation:\n", df_median)
df_median.describe()
```

## Mode Imputation

```python
df_mode = df.copy()

df_mode.fillna({
    "Gender": df["Gender"].mode()[0]
}, inplace=True)

print("After Mode Imputation:\n", df_mode)
print("\nGender frequency counts after imputation:\n", df_mode["Gender"].value_counts())
```
