### Step 1: Load and Explore Dataset
We begin by loading the fitness dataset into Pandas and exploring its structure.
This helps us understand the features before applying scaling.


```python
import pandas as pd

# Load dataset (replace with actual file name if different)
df = pd.read_csv("fitness_activity_100+_preprocess.csv")

# Explore dataset
print("First 5 rows:\n", df.head())
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nStatistical Summary:\n", df.describe())
```

### Step 2: Apply Standard Scaling
StandardScaler transforms features so they have mean = 0 and standard deviation = 1.
This is useful when features are measured in different units but we want them on a comparable scale.


```python
from sklearn.preprocessing import StandardScaler

# Select numeric columns
numeric_cols = ['Steps_Per_Day', 'Calories_Burned', 'Workout_Minutes']

# Apply StandardScaler
scaler = StandardScaler()
df_standard = pd.DataFrame(
    scaler.fit_transform(df[numeric_cols]),
    columns=numeric_cols
)

# Keep User_ID alongside scaled data
df_standard = pd.concat([df['User_ID'], df_standard], axis=1)

print("First 5 rows after Standard Scaling:\n", df_standard.head())
print("\nStatistical Summary:\n", df_standard.describe())
```

### Step 3: Apply Min-Max Scaling
Min-Max Scaler rescales features to a fixed range, usually [0,1].
This is useful when we want to preserve relative relationships but normalize values.


```python
from sklearn.preprocessing import MinMaxScaler

# Select numeric columns
numeric_cols = ['Steps_Per_Day', 'Calories_Burned', 'Workout_Minutes']

# Apply Min-Max Scaler
scaler = MinMaxScaler()
df_minmax = pd.DataFrame(
    scaler.fit_transform(df[numeric_cols]),
    columns=numeric_cols
)

# Keep User_ID alongside scaled data
df_minmax = pd.concat([df['User_ID'], df_minmax], axis=1)

print("First 5 rows after Min-Max Scaling:\n", df_minmax.head())
print("\nStatistical Summary:\n", df_minmax.describe())
```

### Step 4: Comparative Analysis

**StandardScaler**
- Transforms data so each feature has mean = 0 and standard deviation = 1.
- Produces values that can be negative or greater than 1.
- Best suited for algorithms that assume normally distributed data (e.g., Logistic Regression, PCA, Linear Regression).

**Min-Max Scaler**
- Rescales features to a fixed range [0,1].
- Preserves relative relationships between values.
- Best suited for algorithms sensitive to absolute magnitudes (e.g., Neural Networks, KNN, Gradient Descent-based models).

**Observations from our dataset:**
- StandardScaler centered all features around 0 with unit variance. For example, `Steps_Per_Day` now ranges roughly from -1.7 to +1.7.
- Min-Max Scaler compressed all features into [0,1]. For example, `Steps_Per_Day` now ranges strictly between 0 and 1.
- StandardScaler is useful when features have very different variances and we want them comparable.
- Min-Max Scaler is useful when we want bounded values, especially for distance-based or gradient-based algorithms.

**Conclusion:**
Both scaling methods achieve uniformity but in different ways. Choice depends on the algorithm and the nature of the dataset.
