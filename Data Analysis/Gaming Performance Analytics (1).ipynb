## Step 1: Setup Dataset

```python
import numpy as np

data = np.array([
    [501,45,3,120,60],
    [502,30,2,80,50],
    [503,60,4,np.nan,70],
    [504,25,1,60,40],
    [505,80,5,200,90],
    [506,-20,2,70,50],
    [507,35,2,90,55],
    [508,50,3,np.nan,65],
    [509,90,6,250,100],
    [510,200,4,150,80],
    [511,55,3,110,60],
    [512,40,2,95,55]
])

print("Dataset shape:", data.shape)
```

## Step 2: Indexing & Slicing

```python
print("First 5 players:\n", data[:5])
print("Session durations:\n", data[:,1])
print("Single player (ID 505):\n", data[4])
```

## Step 3: Basic Analysis

```python
avg_duration = np.nanmean(data[:,1])   # handle NaN
max_duration = np.nanmax(data[:,1])
min_duration = np.nanmin(data[:,1])
avg_levels = np.nanmean(data[:,2])
total_actions = np.nansum(data[:,3])

print("Average session duration:", avg_duration)
print("Max duration:", max_duration)
print("Min duration:", min_duration)
print("Average levels completed:", avg_levels)
print("Total actions:", total_actions)
```

## Step 4: Vectorization

```python
bonus_actions = np.where(np.isnan(data[:,3]), np.nan, data[:,3] + 10)
high_performers = data[data[:,2] > 4]
low_activity = data[data[:,3] < 80]

print("Bonus actions:", bonus_actions)
print("High performers:\n", high_performers)
print("Low activity players:\n", low_activity)
```

## Step 5: Feature Engineering

```python
performance_score = data[:,2] / np.where(data[:,1] <= 0, np.nan, data[:,1])
data = np.column_stack((data, performance_score))

print("Dataset with Performance Score:\n", data)
```

## Step 6: Data Cleaning

```python
# Replace NaN with mean of column
col_means = np.nanmean(data, axis=0)
inds = np.where(np.isnan(data))
data[inds] = np.take(col_means, inds[1])

# Replace negative values in session_duration with mean
data[:,1] = np.where(data[:,1] < 0, col_means[1], data[:,1])

# Handle outliers (mean ± 2*std)
mean_duration = np.mean(data[:,1])
std_duration = np.std(data[:,1])
upper_limit = mean_duration + 2*std_duration
lower_limit = mean_duration - 2*std_duration
data[:,1] = np.clip(data[:,1], lower_limit, upper_limit)

print("Cleaned dataset:\n", data)
```
